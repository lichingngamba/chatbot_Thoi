```python
from chatterbot import ChatBot
from chatterbot.trainers import ListTrainer
from chatterbot.trainers import ChatterBotCorpusTrainer
```

```python
BankBot = ChatBot(name = 'BankBot',
                  read_only = False,                  
                  logic_adapters = ["chatterbot.logic.BestMatch"],                 
                  storage_adapter = "chatterbot.storage.SQLStorageAdapter")
```

```python
corpus_trainer = ChatterBotCorpusTrainer(BankBot)
```

```python
corpus_trainer.train("chatterbot.corpus.english")
```

```python
greet_conversation = [
    "Hello",
    "Hi there!",
    "How are you doing?",
    "I'm doing great.",
    "That is good to hear",
    "Thank you.",
    "You're welcome."
]
 
conversation = [
    "In which domain i am gonna chose my carrer?",
    "you are gonna chosoe in Data_Mining",
]
 
conversation1= [
    "Are you interested in Phython",
    "Phython is my favorite language",
]

#Initializing Trainer Object
trainer = ListTrainer(BankBot)

#Training BankBot
trainer.train(greet_conversation)
trainer.train(conversation)
trainer.train(conversation1)
```

```python
while (True):
    user_input = input("You:" )
    if (user_input == 'quit'):
        print("Have a nice day")
        break
    response = BankBot.get_response(user_input)
    print("Thoi:", response)
```
